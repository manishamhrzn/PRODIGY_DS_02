Perform data cleaning and exploratory data analysis (EDA) on a dataset of your choice, such as the Titanic dataset from Kaggle. Explore the relationships between variables and identify patterns and trends in the data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d1 = pd.read_csv('/Users/MANISHA/content/tt/train.csv')
d1.head()
```

```python
d1.isnull().sum()
```

```python
d1.drop(['Cabin','PassengerId','Name','Ticket'], axis=1, inplace=True)
d1
```

```python
d1['Embarked'] = d1['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
d1
```

```python
d1['Sex'] = d1['Sex'].map({'male': 0, 'female': 1})
d1
```

```python
d1.describe()
```

```python
sns.heatmap(d1.isnull(),cbar=False,cmap='viridis')
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(d1.corr(),cbar=True,annot=True,cmap='Blues')
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(d1[d1['Survived'] == 1]['Age'], label='Survived', kde=True)
sns.histplot(d1[d1['Survived'] == 0]['Age'], label='Not  Survived', kde=True)
plt.legend()
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=d1)
plt.show()
```

```python
sns.pairplot(d1, hue='Survived')
plt.show()
```
